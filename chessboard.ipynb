### python code for draw chessboard

```python
import numpy as np
import matplotlib.pyplot as plt
def draw_dashboard():
    chessboard=np.zeros((8,8),dtype=int)
    chessboard[1::2, ::2]=1
    chessboard[::2,1::2]=1
    
    plt.figure(figsize=(4,4))
    plt.imshow(chessboard,cmap='gray',interpolation='nearest')
    
    plt.xticks(())
    plt.yticks(())
    plt.title('Cahseboard',fontsize=20)
    
    plt.show()
    
    print("Nothing Much")
    print("Done")
    
draw_dashboard()
```
